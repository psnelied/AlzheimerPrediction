```python
# read train files
X_train = []
y_train = []

classes = ['MildDemented', 'ModerateDemented', 'NonDemented', 'VeryMildDemented']

# you might need to change your path
paths = ["alzheimer_Dataset/train/MildDemented/", 
         "alzheimer_Dataset/train/ModerateDemented/", 
         "alzheimer_Dataset/train/NonDemented/",
         "alzheimer_Dataset/train/VeryMildDemented/"]

for y, path in zip(classes, paths):
  for filename in os.listdir(path):
    a= cv2.imread(path+filename)
    a = cv2.resize(a, (224, 224))
    X_train.append(a)
    y_train.append(y)
```

```python
# read test files
X_test = []
y_test = []

classes = ['MildDemented', 'ModerateDemented', 'NonDemented', 'VeryMildDemented']

# NEEDS TO BE UPDATED: paths of the 4 training folders
paths = ["alzheimer_Dataset/test/MildDemented/", 
         "alzheimer_Dataset/test/ModerateDemented/", 
         "alzheimer_Dataset/test/NonDemented/",
         "alzheimer_Dataset/test/VeryMildDemented/"]

for y, path in zip(classes, paths):
  for filename in os.listdir(path):
    a= cv2.imread(path+filename)
    a = cv2.resize(a, (224, 224))
    X_test.append(a)
    y_test.append(y)
```

# Encode labels 

```python
from tensorflow.keras.utils import to_categorical 

def cat_to_int(a):
    a_ret = [0 for k in range(len(a))]
    for k in range(len(a)):
        if a[k] == 'MildDemented':
            a_ret[k] = 0
        if a[k] == 'ModerateDemented':
            a_ret[k] = 1
        if a[k] == 'NonDemented':
            a_ret[k] = 2
        if a[k] == 'VeryMildDemented':
            a_ret[k] = 3
    return a_ret
```

```python
y_train = cat_to_int(y_train)
y_test = cat_to_int(y_test)
            
y_train = np.array(y_train)
y_test = np.array(y_test)

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

#Custom CNN 

```python
from keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, BatchNormalization, Dropout, MaxPool2D, Flatten, Input


cnn_model = Sequential()
cnn_model.add(Input(shape=(224,224,3)))
cnn_model.add(Conv2D(16, 3, activation='relu', padding='same'))
cnn_model.add(Conv2D(16, 3, activation='relu', padding='same'))
cnn_model.add(MaxPool2D())
cnn_model.add(Conv2D(32, 3, activation='relu', padding='same'))
cnn_model.add(Conv2D(32, 3, activation='relu', padding='same'))
cnn_model.add(BatchNormalization())
cnn_model.add(MaxPool2D())
cnn_model.add(Conv2D(64, 3, activation='relu', padding='same'))
cnn_model.add(Conv2D(64, 3, activation='relu', padding='same'))
cnn_model.add(BatchNormalization())
cnn_model.add(MaxPool2D())        
cnn_model.add(Conv2D(128, 3, activation='relu', padding='same'))
cnn_model.add(Conv2D(128, 3, activation='relu', padding='same'))
cnn_model.add(BatchNormalization())
cnn_model.add(MaxPool2D())        
cnn_model.add(Dropout(0.2))
cnn_model.add(Conv2D(256, 3, activation='relu', padding='same'))
cnn_model.add(Conv2D(256, 3, activation='relu', padding='same'))
cnn_model.add(BatchNormalization())
cnn_model.add(MaxPool2D())        
cnn_model.add(Dropout(0.2))
cnn_model.add(Flatten())
cnn_model.add(Dense(512, activation='relu'))
cnn_model.add(BatchNormalization())
cnn_model.add(Dropout(0.7))        
cnn_model.add(Dense(128, activation='relu'))
cnn_model.add(BatchNormalization())
cnn_model.add(Dropout(0.5))        
cnn_model.add(Dense(64, activation='relu'))
cnn_model.add(BatchNormalization())
cnn_model.add(Dropout(0.3))        
cnn_model.add(Dense(4, activation='softmax'))

cnn_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
cnn_model.summary()
```

## Training with SMOTE

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE 

X_train_flatten = X_train.reshape(5121, 150528)
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_flatten, y_train)

X_train_smote = X_train_smote.reshape(10240,224,224,3)
y_train_smote = y_train_smote

X_train_smote_ = X_train_smote/255.
X_test_ = X_test/255.

X_train_vision, X_val_vision, y_train_vision, y_val_vision= train_test_split(X_train_smote_, y_train_smote, 
                                                                 test_size = 0.2, random_state=42)





reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor = 'val_accuracy',
                                                 factor = 0.2,
                                                 patience = 2,
                                                 verbose = 1,
                                                 min_delta = 1e-4,
                                                 min_lr = 1e-6,
                                                 mode = 'max')

earlystopping = tf.keras.callbacks.EarlyStopping(monitor = 'val_accuracy',
                                                 min_delta = 1e-4,
                                                 patience = 5,
                                                 mode = 'max',
                                                 restore_best_weights = True,
                                                 verbose = 1)

checkpointer = tf.keras.callbacks.ModelCheckpoint(filepath = './model.hdf5',
                                                  monitor = 'val_accuracy', 
                                                  verbose = 1, 
                                                  save_best_only = True,
                                                  save_weights_only = True,
                                                  mode = 'max')

callbacks = [earlystopping, reduce_lr, checkpointer]

history_vision = model.fit(x = X_train_vision, y = y_train_vision ,validation_data = (X_val_vision, y_val_vision),epochs = 5,callbacks=callbacks)
print("Testing Accuracy: %.2f%%"%( model.evaluate(X_test_, y_test)[1] * 100))

```
