# Read in train and test set


```python
import cv2
import os
from matplotlib import pyplot as plt
```

```python
# read train files
X_train = []
y_train = []


classes = ['MildDemented', 'ModerateDemented', 'NonDemented', 'VeryMildDemented']

# you might need to change your path
paths = ["/content/ddd/Alzheimer_s Dataset/train/MildDemented/", 
         "/content/ddd/Alzheimer_s Dataset/train/ModerateDemented/", 
         "/content/ddd/Alzheimer_s Dataset/train/NonDemented/",
         "/content/ddd/Alzheimer_s Dataset/train/VeryMildDemented/"]

for y, path in zip(classes, paths):
  for filename in os.listdir(path):
    a= cv2.imread(path+filename, -1)
    X_train.append(a)
    y_train.append(y)
```

```python
# read test files
X_test = []
y_test = []

classes = ['MildDemented', 'ModerateDemented', 'NonDemented', 'VeryMildDemented']

# NEEDS TO BE UPDATED: paths of the 4 training folders
paths = ["/content/ddd/Alzheimer_s Dataset/test/MildDemented/", 
         "/content/ddd/Alzheimer_s Dataset/test/ModerateDemented/", 
         "/content/ddd/Alzheimer_s Dataset/test/NonDemented/",
         "/content/ddd/Alzheimer_s Dataset/test/VeryMildDemented/"]

for y, path in zip(classes, paths):
  for filename in os.listdir(path):
    a= cv2.imread(path+filename, -1)
    X_test.append(a)
    y_test.append(y)
```

# Resampling

### flatten (or other reshaping?)

```python
# flatten the images
X_train_flat = [i.reshape(1,-1).flatten() for i in X_train]
X_test_flat = [i.reshape(1,-1).flatten() for i in X_test]
```

```python
# number of columns in a flattened image
n_features = X_train_flat[10].reshape(1, -1).shape[1]
n_features
```

# Vanilla Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression(multi_class='multinomial')
logisticRegr.fit(X_train_flat, y_train)

```

```python
score = logisticRegr.score(X_train_flat, y_train)
print('train score:',score)

score = logisticRegr.score(X_test_flat, y_test)
print('test score:',score)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics

predictions = logisticRegr.predict(X_test_flat)

cm = metrics.confusion_matrix(y_test, predictions)
print(cm)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```

# SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_flat, y_train)
```

```python
from collections import Counter

print(f'shape of development dataset: {len(X_train_smote)}\n')
Counter(y_train_smote)
```

```python
solvers=['lbfgs', 'sag', 'saga','newton-cg']
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LogisticRegression

all_scores = []
for i in solvers:
  logreg =  LogisticRegression('l2', multi_class='multinomial', solver = i)

  scores = cross_val_score(logreg, X_train_smote, y_train_smote, cv=5, error_score='raise')

  print(np.mean(scores))
  all_scores.append([i, np.mean(scores)])
```

```python
print("scores:",all_scores)
```

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression('l2', multi_class='multinomial', solver='newton-cg', max_iter=100)
```

```python
logisticRegr.fit(X_train_smote, y_train_smote)
```

```python
score = logisticRegr.score(X_train_smote, y_train_smote)
print('test score:',score)

score = logisticRegr.score(X_test_flat, y_test)
print('test score:',score)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics

predictions = logisticRegr.predict(X_test_flat)

cm = metrics.confusion_matrix(y_test, predictions)
print(cm)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```
